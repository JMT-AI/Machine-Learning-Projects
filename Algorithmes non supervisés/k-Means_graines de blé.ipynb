Il s'agit de trier en clusters un jeu de données comtenant différentes catégories de graines de blé en fonction de leurs caractéristiques via un algorithme non supervisé k-means et de l'évaluer.

```python
# Importations

import pandas as pd
%pylab inline

from sklearn.preprocessing import StandardScaler
from sklearn import cluster, metrics, decomposition
```

```python
# recuparation

data = pd.read_csv('seeds.txt', delim_whitespace=True, header=None)
data.head()
```

```python
X = data.drop(7, axis=1).values
Y = data[7].values
```

```python
X.shape, Y.shape
```

```python
scale = StandardScaler()
X_scaled = scale.fit_transform(X)
```

```python
# Afficher le coef de silhouettes en fonction du nombre de clusters

silhouettes = []

for num_clusters in range(2, 10):
    cls = cluster.KMeans(n_clusters=num_clusters, n_init=1, init='random')
    cls.fit(X_scaled)
    silh = metrics.silhouette_score(X_scaled, cls.labels_)
    silhouettes.append(silh)

# representation
plt.plot(range(2,10), silhouettes, marker='o')
```

On remarque que 2 ou 3 clusters optimise le coefficient de silhouette. 
Explorons les données en 2 dimensions via PCA :

```python
# projection en 2 dimensions via PCA
pca = decomposition.PCA(n_components=2)

pca.fit(X_scaled)
```

```python
print("La 1ere composante explique {:.2f} % de la variance et la 2e {:.2f} % \
de la variance".format(pca.explained_variance_ratio_[0]*100, pca.explained_variance_ratio_[1]*100))
      
print("les 2 composantes expliquent {:.2f} % de la variance".format(pca.explained_variance_ratio_.cumsum()[1]*100))
```

```python
X_trans = pca.transform(X_scaled)
```

```python
# Check 2 dimensions
X_trans.shape
```

```python
# representation des clusters

fig = plt.figure(figsize=(10,5))

# avec 2 clusters
cls2 = cluster.KMeans(n_clusters=2)
cls2.fit(X_scaled)
ax = fig.add_subplot(121)
ax.scatter(X_trans[:, 0], X_trans[:, 1], c=cls2.labels_)

# avec 3 clusters
cls3 = cluster.KMeans(n_clusters=3)
cls3.fit(X_scaled)
ax = fig.add_subplot(122)
ax.scatter(X_trans[:, 0], X_trans[:, 1], c=cls3.labels_)
```

```python
# Comparaison avec les labels réels

fig = plt.figure(figsize=(10,5))

# affichage des data embbeded clusterisés par les labels trouvés par cls3
cls3 = cluster.KMeans(n_clusters=3)
cls3.fit(X_scaled)
ax = fig.add_subplot(121)
ax.scatter(X_trans[:, 0], X_trans[:, 1], c=cls3.labels_)
ax.set_title('algorithme k-Means')


# affichage des data embbeded clusterisés par les labels réels Y
ax = fig.add_subplot(122)
ax.scatter(X_trans[:, 0], X_trans[:, 1], c=Y)
ax.set_title('étiquettes réelles')

plt.show()
```

#### Quantifions la correspondance entre les étiquettes trouvées par notre cluster et les vraies étiquettes :

```python
print(metrics.adjusted_rand_score(Y, cls3.labels_))
```

#### Plus adjusted_rand_score est proche de 1, mieux l'algorithme de clustering correspond au étiquettes initiales.
