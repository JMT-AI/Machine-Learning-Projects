```python
# Importations

import pandas as pd
%pylab inline

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn import kernel_ridge
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

### 1. Preparation et exploration

```python
# recuperation des donnees
data_row = pd.read_csv('winequality-white.csv', sep=';')
data_row.head()
```

### 2. Entrainement d'une KRR  (regression ridge à noyau)

Nous pouvons maintenant entraîner une kRR sur le jeu d'entraînement en utilisant des paramètres par défaut pour le paramètre de régularisation et la bande passante du noyau RBF gaussien :

```python
# Création d'une KRR avec un paramètre de régularisation et une bande passante du noyau gaussien par défaults

predictor = kernel_ridge.KernelRidge(
    alpha=1.0,    # valeur par défaut 
    kernel='rbf', # noyau Gaussien
    gamma=0.01)   # valeur de 1/(2 * sigma**2))

# entrainement du prédicteur sur le jeu d'entrainement
predictor.fit(X_train, y_train)

# predictions sur le jeu de test
y_pred = predictor.predict(X_test)

# calculer la RMSE sur le jeu de test
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print("RMSE = %.2f" % rmse)
```

On obtient une RMSE de 0.72, ce qui est correct pour des nombres entiers allant de 3 à 9.

Pour visualiser les prédictions, on peut utiliser un nuage de points dans lequel la surface de chaque point est proportionnelle au nombre d'observations ayant exactement ces valeurs de score prédit et de score réel.

```python
# créer une figure
fig = plt.figure(figsize=(6, 6))

# Compter, pour chaque paire de valeurs (y, y') où y est un vrai score et y' le score prédit,
# le nombre de ces paires.
# Ce nombre sera utilisé pour modifier la taille des marqueurs correspondants dans un nuage de points

pairs_and_sizes = {}

for (yt, yp) in zip(list(y_test), list(y_pred)):
    if (yt, yp) in pairs_and_sizes.keys():
        pairs_and_sizes[(yt, yp)] += 10
    else:
        pairs_and_sizes[(yt, yp)] = 10

# on recupère toutes les uniques clés du dict 'pairs_and_sizes'
keys = pairs_and_sizes.keys()

tailles = [pairs_and_sizes[k] for k in keys]

# afficher les prédictions
plt.scatter([k[0] for k in keys], [k[1] for k in keys], s=tailles, label="gamma = 0.01: RMSE = {:.2f}".format(rmse))

# étiqueter les axes et le graphique
plt.xlabel('Vrai score', fontsize=16)
plt.ylabel(u'Score prédit', fontsize=16)
plt.title('kernel Ridge Regression', fontsize=16)

# limites des axes
plt.xlim([2.9, 9.1])
plt.ylim([2.9, 9.1])

# afficher la légende
plt.legend(loc="lower right", fontsize=12)
plt.show()
```

On observe une corrélation entre les scores prédits et les score réels.

Comme dans le cas de la SVM à noyau, le paramètre gamma du noyau RBF gaussien joue un rôle important : s'il est trop élevé, la matrice de Gram sur le jeu d'entraînement est dominée par sa diagonale et la kRR ne peut pas apprendre. Si l'on recommence avec gamma=50, on obtient les prédictions suivantes :

```python
predictor_g50 = kernel_ridge.KernelRidge(
    alpha=1.0,    # valeur par défaut 
    kernel='rbf', # noyau Gaussien
    gamma=50)   # valeur de 1/(2 * sigma**2))

# entrainement du prédicteur sur le jeu d'entrainement
predictor_g50.fit(X_train, y_train)

# predictions sur le jeu de test
y_pred_g50 = predictor_g50.predict(X_test)

# calculer la RMSE sur le jeu de test
rmse_g50 = np.sqrt(metrics.mean_squared_error(y_test, y_pred_g50))
print("RMSE = %.2f  --> Very bad with rbf bandwith = 50" % rmse_g50)
```

```python
# créer une figure
fig = plt.figure(figsize=(6, 6))

# Compter, pour chaque paire de valeurs (y, y') où y est un vrai score et y' le score prédit,
# le nombre de ces paires.
# Ce nombre sera utilisé pour modifier la taille des marqueurs correspondants dans un nuage de points

pairs_and_sizes_g50 = {}

for (yt, yp) in zip(list(y_test), list(y_pred_g50)):
    if (yt, yp) in pairs_and_sizes_g50.keys():
        pairs_and_sizes_g50[(yt, yp)] += 10
    else:
        pairs_and_sizes_g50[(yt, yp)] = 10

# on recupère toutes les uniques clés du dict 'pairs_and_sizes_g50'
keys_g50 = pairs_and_sizes_g50.keys()

tailles_g50 = [pairs_and_sizes_g50[k] for k in keys_g50]

# afficher les prédictions
plt.scatter([k[0] for k in keys_g50], [k[1] for k in keys_g50], s=tailles_g50, label="gamma = 0.01: RMSE = {:.2f}".format(rmse_g50))

# on affiche aussi le nuage pour gamma=0.01
plt.scatter([k[0] for k in keys], [k[1] for k in keys], s=tailles, label="gamma = 0.01: RMSE = {:.2f}".format(rmse))

# étiqueter les axes et le graphique
plt.xlabel('Vrai score', fontsize=16)
plt.ylabel(u'Score prédit', fontsize=16)
plt.title('kernel Ridge Regression', fontsize=16)

# limites des axes
plt.xlim([0, 9])
plt.ylim([0, 9])

# afficher la légende
plt.legend(loc="upper right", fontsize=12)
plt.show()
```

Pour optimiser le paramètre de régularisation alpha et le paramètre gamma du noyau gaussien, on peut utiliser une recherche sur grille :

```python
# valeurs du paramètre C
alpha_range = np.logspace(-2, 2, 5)

# valeurs du paramètre gamma
gamma_range = np.logspace(-2, 1, 4)

# grille de paramètres
param_grid = {'alpha': alpha_range, 'gamma': gamma_range}

# score pour sélectionner le modèle optimal
score = 'neg_mean_squared_error'

predictor_grid = GridSearchCV(
    kernel_ridge.KernelRidge(kernel='rbf'),
    param_grid,
    cv=5,
    scoring=score )

# entrainement du prédicteur sur le jeu d'entrainement
predictor_grid.fit(X_train, y_train)

# predictions sur le jeu de test
y_pred_grid = predictor_grid.predict(X_test)

# calculer la RMSE sur le jeu de test
rmse_grid = np.sqrt(metrics.mean_squared_error(y_test, y_pred_grid))
print("RMSE = %.2f  --> Best performance with rbf cross validation !" % rmse_grid)
```

```python
# figure
fig = plt.figure(figsize=(7, 7))

# Compter, pour chaque paire de valeurs (y, y') où y est un vrai score et y' le score prédit,
# le nombre de ces paires.
# Ce nombre sera utilisé pour modifier la taille des marqueurs correspondants dans un nuage de points

pairs_and_sizes_grid = {}

for (yt, yp) in zip(list(y_test), list(y_pred_grid)):
    if (yt, yp) in pairs_and_sizes_grid.keys():
        pairs_and_sizes_grid[(yt, yp)] += 15
    else:
        pairs_and_sizes_grid[(yt, yp)] = 15

# on recupère toutes les uniques clés du dict 'pairs_and_sizes_g50'
keys_grid = pairs_and_sizes_grid.keys()

tailles_grid = [pairs_and_sizes_grid[k] for k in keys_grid]

# afficher les prédictions
plt.scatter([k[0] for k in keys_grid], [k[1] for k in keys_grid], s=tailles_grid, label="gamma = 0.01: RMSE = {:.2f}".format(rmse_grid))

# on affiche aussi le nuage pour alpha et gamma par défaut
plt.scatter([k[0] for k in keys], [k[1] for k in keys], s=tailles, label="gamma = 0.01: RMSE = {:.2f}".format(rmse))

# étiqueter les axes et le graphique
plt.xlabel('Vrai score', fontsize=16)
plt.ylabel(u'Score prédit', fontsize=16)
plt.title('kernel Ridge Regression', fontsize=16)

# limites des axes
plt.xlim([0, 9])
plt.ylim([0, 9])

# afficher la légende
plt.legend(loc="upper right", fontsize=12)
plt.show()
```

On obtient une RMSE légèrement meilleure (0.71) en optimisant le modèle.

Les prédictions faites se superposent à peu près à celles obtenues avec gamma=0.01
